```python
# Time-Series Data Analytics using Pandas
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# 1 Create sample time-series data (Date vs Temperature)
dates = pd.date_range(start="2024-01-01", periods=30, freq="D")
temps = [28,29,30,32,31,29,28,30,33,32,31,30,29,28,27,29,30,32,31,33,34,35,33,32,30,31,29,28,27,26]
data = pd.DataFrame({"Date": dates, "Temperature": temps})
data.set_index("Date", inplace=True)

# 2 Plot the time-series
data.plot(title="Daily Temperature Trend", marker='o')
plt.ylabel("Temperature (C)")
plt.show()

# 3 Calculate moving average (trend smoothing)
data["Moving_Avg"] = data["Temperature"].rolling(window=5).mean()
data[["Temperature", "Moving_Avg"]].plot(title="Temperature with Moving Average")
plt.show()

# 4 Simple ARIMA model for forecasting
model = ARIMA(data["Temperature"], order=(1,1,1))
fit = model.fit()
forecast = fit.forecast(steps=5)
print("Next 5-day Temperature Forecast:\n", forecast)
```
