```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
edges = [
    ("Alice", "Bob"),
    ("Alice", "David"),
    ("Bob", "Charlie"),
    ("Charlie", "David"),
    ("David", "Eve"),
    ("Eve", "Frank"),
    ("Frank", "George")
]

G.add_edges_from(edges)
nx.draw(G, with_labels=True,node_size=900, font_size=10)
plt.title("Social Network Graph")
plt.show()

print("Nodes:", G.nodes())
print("Edges:", G.edges())
print("degree of each Node:",dict(G.degree()))

centrality = nx.degree_centrality(G)
print("Degree Centrality:", centrality)

print("Shortest path between Alice and George:", nx.shortest_path(G, "Alice", "George"))

# 6️⃣ Community detection (using connected components)
communities = list(nx.connected_components(G))
print("Detected Communities:", communities)


```
