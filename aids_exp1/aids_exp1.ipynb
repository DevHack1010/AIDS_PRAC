```

import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp('Apple is looking at buying U.K. startup for $1 billion')
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop)
```

```

import nltk
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer
text = 'Apple is looking at buying U.K. startup for $1 billion'
analyzer = SentimentIntensityAnalyzer()
scores = analyzer.polarity_scores(text)
print(scores)
```

```

import torch

def compute_fourier_basis(n=100):
    J = torch.zeros((3,n))
    for i in range(n):
        J[0][i] = 0.5
        J[1][i] = torch.cos(torch.tensor(i*3.14/n))
        J[2][i] = torch.sin(torch.tensor(i*3.14/n))
    return J

J = compute_fourier_basis()
print(J)
```

```

import torch
import torch.nn as nn

class EmbeddingExample(nn.Module):
    def __init__(self):
        super(EmbeddingExample, self).__init__()
        self.embedding = nn.Embedding(10, 3)

    def forward(self, x):
        return self.embedding(x)

embedding_example = EmbeddingExample()
input_indices = torch.LongTensor([1,2,4,5])
output = embedding_example(input_indices)
print(output)
```
