```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt


# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]


# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)

# Creating color parameters
colors = ("orange", "cyan", "brown",
          "grey", "indigo", "beige")

# Wedge properties
wp = {'linewidth': 1, 'edgecolor': "green"}

# Creating autocpt arguments


def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)


# Creating plot
fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  autopct=lambda pct: func(pct, data),
                                  explode=explode,
                                  labels=cars,
                                  shadow=True,
                                  colors=colors,
                                  startangle=90,
                                  wedgeprops=wp,
                                  textprops=dict(color="magenta"))

# Adding legend
ax.legend(wedges, cars,
          title="Cars",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")
ax.set_title("Customizing pie chart")

# show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('GlobalWeatherRepository.csv')


df['last_updated'] = pd.to_datetime(df['last_updated'])


city_df = df[df['location_name'] == 'New Delhi']


city_df['Month'] = city_df['last_updated'].dt.month
monthly_avg = city_df.groupby('Month')['temperature_celsius'].mean()


warmest_month = monthly_avg.idxmax()
coldest_month = monthly_avg.idxmin()

# Plotting
plt.figure(figsize=(12,6))
plt.plot(monthly_avg.index, monthly_avg.values, linestyle='-', marker='o', color='b')

# Highlight warmest and coldest months
plt.scatter(warmest_month, monthly_avg[warmest_month], color='r', label='Warmest Month')
plt.scatter(coldest_month, monthly_avg[coldest_month], color='c', label='Coldest Month')

# After plotting scatter points
plt.text(warmest_month, monthly_avg[warmest_month]+0.5,
         f'Warmest: {warmest_month}\n{monthly_avg[warmest_month]:.1f}C',
         ha='center', color='red')

plt.text(coldest_month, monthly_avg[coldest_month]-1,
         f'Coldest: {coldest_month}\n{monthly_avg[coldest_month]:.1f}C',
         ha='center', color='cyan')


# Add labels
plt.xlabel('Month')
plt.ylabel('Average Temperature (C)')
plt.title('Monthly Average Temperature Trend - New Delhi')
plt.xticks(monthly_avg.index)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#q2.1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
channels = ['Social Media', 'Google Ads', 'Email', 'Referrals']
data = {
    'SignUp_ID': range(1, 31),
    'Channel': np.random.choice(channels, size=30)
}
df = pd.DataFrame(data)


print(df.head())
```

```python
channel_counts = df['Channel'].value_counts()

plt.figure(figsize=(8,6))
bars = plt.bar(channel_counts.index, channel_counts.values, color='skyblue', edgecolor='black')

plt.xlabel('Advertising Channel')
plt.ylabel('Number of Sign-ups')
plt.title('New Customer Sign-ups per Channel (Last Quarter)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

channel_counts_sorted = channel_counts.sort_values(ascending=False)

plt.figure(figsize=(8,6))
bars = plt.bar(channel_counts_sorted.index, channel_counts_sorted.values, color='lightgreen', edgecolor='black')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f'{int(height)}', ha='center', va='bottom')

plt.xlabel('Advertising Channel')
plt.ylabel('Number of Sign-ups')
plt.title('New Customer Sign-ups per Channel (Sorted)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

last_quarter = {'Social Media': 8, 'Google Ads': 10, 'Email': 7, 'Referrals': 5}
prev_quarter = {'Social Media': 6, 'Google Ads': 9, 'Email': 5, 'Referrals': 8}

channels = list(last_quarter.keys())
last_counts = list(last_quarter.values())
prev_counts = list(prev_quarter.values())

x = np.arange(len(channels))
bar_width = 0.35

plt.figure(figsize=(10,6))
plt.bar(x - bar_width/2, last_counts, width=bar_width, label='Last Quarter', color='cornflowerblue', edgecolor='black')
plt.bar(x + bar_width/2, prev_counts, width=bar_width, label='Previous Quarter', color='salmon', edgecolor='black')

plt.xlabel('Advertising Channel')
plt.ylabel('Number of Sign-ups')
plt.title('Customer Sign-ups Comparison Across Two Quarters')
plt.xticks(x, channels)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# q3.1

import pandas as pd
import numpy as np

np.random.seed(0)
companies = [f'Company_{i}' for i in range(1, 31)]
market_shares = np.random.randint(1, 10, size=30)

market_df = pd.DataFrame({
    'Company': companies,
    'MarketShare': market_shares
})


print(market_df.head())
```

```python
import matplotlib.pyplot as plt

total_share = market_df['MarketShare'].sum()
market_df['Percentage'] = (market_df['MarketShare'] / total_share) * 100


main_companies = market_df[market_df['Percentage'] >= 3].copy()
others = market_df[market_df['Percentage'] < 3].copy()

others_sum = others['MarketShare'].sum()
others_row = pd.DataFrame({
    'Company': ['Others'],
    'MarketShare': [others_sum],
    'Percentage': [(others_sum / total_share) * 100]
})

final_df = pd.concat([main_companies, others_row], ignore_index=True)

explode = [0.1 if c == 'Company_5' else 0 for c in final_df['Company']]


plt.figure(figsize=(8,8))
plt.pie(final_df['MarketShare'], labels=final_df['Company'], autopct='%1.1f%%', startangle=140, shadow=True)
plt.title('Market Share Distribution by Company')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Seed for reproducibility
np.random.seed(42)

# Generate sample data for 50 experimental plots
n = 50
fertilizer = np.random.uniform(20, 100, n)  # fertilizer amount in kg
# Assume yield roughly increases with fertilizer with some noise
crop_yield = 1.2 * fertilizer + np.random.normal(0, 10, n)

# Add a categorical variable for soil type (A, B, C)
soil_types = np.random.choice(['A', 'B', 'C'], n)

# Create DataFrame
df = pd.DataFrame({
    'Fertilizer (kg)': fertilizer,
    'Crop Yield (bushels)': crop_yield,
    'Soil Type': soil_types
})
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Fertilizer (kg)'], df['Crop Yield (bushels)'], alpha=0.8)
plt.xlabel('Fertilizer Used (kg)')
plt.ylabel('Crop Yield (bushels)')
plt.title('Scatter Plot: Fertilizer vs. Crop Yield')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='Fertilizer (kg)',
    y='Crop Yield (bushels)',
    hue='Soil Type',
    data=df,
    palette='Set1',
    s=80
)
plt.title('Scatter Plot with Soil Type as Color')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='Fertilizer (kg)',
    y='Crop Yield (bushels)',
    hue='Soil Type',
    data=df,
    palette='Set1',
    s=100,
    alpha=0.6
)
plt.title('Scatter Plot with Adjusted Size and Transparency')
plt.grid(True)
plt.show()
```

```python
# q6

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


np.random.seed(42)
spending = np.random.normal(loc=250, scale=75, size=50)
spending = np.round(spending, 2)

df = pd.DataFrame({'CustomerID': range(1, 51), 'Spending': spending})
print(df.head())
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Spending'], bins=10, edgecolor='black')
plt.title('Customer Spending Distribution')
plt.xlabel('Spending Amount ($)')
plt.ylabel('Number of Customers')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Spending'], bins=10, density=True, edgecolor='black')
plt.title('Customer Spending Density Distribution')
plt.xlabel('Spending Amount ($)')
plt.ylabel('Probability Density')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
np.random.seed(42)
df['Gender'] = np.random.choice(['Male', 'Female'], size=50)

male_spending = df[df['Gender']=='Male']['Spending']
female_spending = df[df['Gender']=='Female']['Spending']

plt.figure(figsize=(10,6))
plt.hist(male_spending, bins=10, alpha=0.7, label='Male', color='blue', edgecolor='black')
plt.hist(female_spending, bins=10, alpha=0.7, label='Female', color='pink', edgecolor='black')
plt.title('Spending Distribution by Gender')
plt.xlabel('Spending Amount ($)')
plt.ylabel('Number of Customers')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
